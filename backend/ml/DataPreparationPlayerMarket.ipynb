```python
#standard feedforward network

import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd

# Load dataset
#df = pd.read_csv("D:\PIDS\dataset\player_dataset_cleaned.csv")
df = df_clean.drop(columns=["player", "PlayerId"])  # Removing text identifiers

# Define features and target
important_features = ["age", "performance_metrics", "media_sentiment"]
feature_weights = np.ones(df.shape[1] - 1)  # Default weight = 1
for feature in important_features:
    if feature in df.columns:
        feature_weights[df.columns.get_loc(feature) - 1] = 1.5  # Increase importance
# Define features and target
X = df.drop(columns=["MarketValue"]).values #* feature_weights
y = df["MarketValue"].values.reshape(-1, 1)  # Ensure y is 2D for inverse scaling

# Define Quantile Regression Model
class MarketValuePredictor(nn.Module):
    def __init__(self, input_dim, range_percent=0.1):  # 10% by default
        super(MarketValuePredictor, self).__init__()
        self.range_percent = range_percent

        self.network = nn.Sequential(
            nn.Linear(input_dim, 256),
            nn.LeakyReLU(),
            nn.BatchNorm1d(256),
            nn.Dropout(0.3),
            
            nn.Linear(256, 128),
            nn.LeakyReLU(),
            nn.BatchNorm1d(128),
            nn.Dropout(0.3),
            
            nn.Linear(128, 64),
            nn.LeakyReLU(),
            nn.BatchNorm1d(64),
            nn.Dropout(0.2),

            nn.Linear(64, 1)  # Single output: Predicted Market Value
        )

    #def forward(self, x):
    #    return self.model(x)
    def forward(self, x):
        center = self.network(x).squeeze()
        half_range = center * (self.range_percent / 2)

        lower = center - half_range
        upper = center + half_range

        return torch.stack([lower, upper], dim=1)

# K-Fold Cross-Validation
k_folds = 5
kf = KFold(n_splits=k_folds, shuffle=True, random_state=42)
fold_rmse_scores = []

for fold, (train_idx, val_idx) in enumerate(kf.split(X)):
    print(f"Training fold {fold+1}/{k_folds}...")

    # Split data for this fold
    X_train, X_val = X[train_idx], X[val_idx]
    y_train, y_val = y[train_idx], y[val_idx]

    # Convert to PyTorch tensors
    X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
    X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
    y_train_tensor = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
    y_val_tensor = torch.tensor(y_val, dtype=torch.float32).view(-1, 1)

    # Initialize model
    input_dim = X_train.shape[1]
    model = MarketValuePredictor(input_dim, range_percent=0.2)  # 50% range

    # Optimizer & Scheduler
    optimizer = optim.AdamW(model.parameters(), lr=0.0005, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=10, verbose=True)

    # Training loop
    num_epochs = 300
    early_stopping_patience = 20
    best_loss = float('inf')
    stopping_counter = 0

    for epoch in range(num_epochs):
        model.train()
        optimizer.zero_grad()
        predictions = model(X_train_tensor)

        # Compute Mean Squared Error loss
        
        #loss = nn.MSELoss()(predictions, y_train_tensor)
        loss_lower = quantile_loss(predictions[:, 0], y_train_tensor.squeeze(), 0.1)
        loss_upper = quantile_loss(predictions[:, 1], y_train_tensor.squeeze(), 0.9)
        #loss = loss_lower + loss_upper
        
        lambda_penalty = 0.1  # Increase lambda to emphasize strict range control
        loss = loss_lower + loss_upper + lambda_penalty
        
        loss.backward()
        optimizer.step()
        scheduler.step(loss)

        if loss.item() < best_loss:
            best_loss = loss.item()
            stopping_counter = 0
        else:
            stopping_counter += 1
            if stopping_counter >= early_stopping_patience:
                print(f"Early stopping at epoch {epoch+1}")
                break

        if (epoch + 1) % 10 == 0:
            print(f"Fold {fold+1}, Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")

    # Evaluate on validation set
    model.eval()
    y_pred_tensor = model(X_val_tensor)
    y_pred_scaled = y_pred_tensor.detach().numpy()
    
    # Reverse Standardization (Inverse Transform)
    # Create a DataFrame with the same structure as the original standardized data
    y_pred_original = pd.DataFrame(y_pred_scaled, columns=["MarketValue_Lower", "MarketValue_Upper"])  # Assuming the model predicts only 2 columns
    y_val_original = pd.DataFrame(y_val, columns=["MarketValue"])  # Assuming y_val corresponds to "MarketValue"

    # Apply inverse transform only to the standardized columns
    y_pred_original["MarketValue_Lower"] = scaler_market_value.inverse_transform(y_pred_original[["MarketValue_Lower"]])
    y_pred_original["MarketValue_Upper"] = scaler_market_value.inverse_transform(y_pred_original[["MarketValue_Upper"]])
    y_val_original["MarketValue"] = scaler_market_value.inverse_transform(y_val_original["MarketValue"].values.reshape(-1, 1))
    
    # Compute RMSE for center estimate
    y_center = (y_pred_original["MarketValue_Lower"] + y_pred_original["MarketValue_Upper"]) / 2
    rmse = np.sqrt(mean_squared_error(y_val_original, y_center))
    fold_rmse_scores.append(rmse)
    print(f"Fold {fold+1} RMSE: {rmse:.4f}")
    
    # Calculate whether the actual values are within the predicted range
    within_range = (y_val_original["MarketValue"] >= y_pred_original["MarketValue_Lower"]) & \
               (y_val_original["MarketValue"] <= y_pred_original["MarketValue_Upper"])
    
    # Calculate accuracy of predictions being within the range
    accuracy = np.mean(within_range) * 100  # Percentage of values within the predicted range
    print(f"Fold {fold+1} Accuracy: {accuracy:.2f}%")
    
    # Print sample predictions
    for i in range(5):
        print(f"Predicted Range: {y_pred_original.iloc[i, 0]:.2f} - {y_pred_original.iloc[i, 1]:.2f}, Actual: {y_val_original.iloc[i, 0]:.2f}")

# Average RMSE across folds
mean_rmse = np.mean(fold_rmse_scores)
print(f"Mean RMSE across {k_folds} folds: {mean_rmse:.4f}")
```
